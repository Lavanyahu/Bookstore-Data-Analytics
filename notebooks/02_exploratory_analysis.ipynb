```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
books = pd.read_csv("../data/books.csv")
users = pd.read_csv("../data/users.csv")
orders = pd.read_csv("../data/orders.csv")
ratings = pd.read_csv("../data/ratings.csv")
```

```python
orders_books = orders.merge(books, on="book_id")
orders_books["revenue"] = orders_books["price"] * orders_books["quantity"]
```

```python
orders_books.head()
```

```python
orders_books = orders.merge(books, on="book_id")
orders_books["revenue"] = orders_books["price"] * orders_books["quantity"]
```

```python
revenue_by_genre = orders_books.groupby("genre")["revenue"].sum()
revenue_by_genre
```

```python
revenue_by_genre.plot(kind="bar", title="Revenue by Genre")
plt.xlabel("Genre")
plt.ylabel("Revenue")
plt.show()
```

```python
top_books = orders_books.groupby("title")["quantity"].sum().sort_values(ascending=False)
top_books
```

```python
top_books.plot(kind="bar", title="Top Selling Books")
plt.xlabel("Book Title")
plt.ylabel("Total Quantity Sold")
plt.show()
```

```python
ratings_books = ratings.merge(books, on="book_id")
avg_rating = ratings_books.groupby("title")["rating"].mean()
avg_rating
```

```python
avg_rating.plot(kind="bar", title="Average Rating per Book")
plt.xlabel("Book Title")
plt.ylabel("Average Rating")
plt.show()
```

```python
## ðŸ” Key Insights

- Fiction and Classics genres generate the highest revenue among all categories.
- Certain books receive consistently high average ratings, indicating strong customer satisfaction.
- A small group of users contributes a large portion of total revenue, showing the presence of high-value customers.
- Genres with higher ratings generally correlate with higher sales volume.
- Revenue distribution suggests opportunities for targeted marketing and inventory optimization.
```
